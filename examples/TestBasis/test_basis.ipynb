```python
from simulai.math.basis import LegendreBasis
from scipy.special import legendre
import numpy as np
import matplotlib.pyplot as plt
```

```python
Omega = np.linspace(-1,1,100)
```

```python
leg = LegendreBasis(10)
```

```python
leg.basis(0)(Omega).shape
legendre(0)(Omega).shape
```

```python
for i in range(4):
    plt.plot(leg.basis(i)(Omega), c='r')
    plt.plot(legendre(i)(Omega), c='b', ls='--')
```

```python
for i in range(1):
    plt.plot(leg.basis(i).deriv()(Omega), c='r')
    plt.plot(legendre(i).deriv()(Omega), c='b', ls='--')
```
